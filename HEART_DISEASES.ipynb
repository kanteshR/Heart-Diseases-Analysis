For better understanding we will change the numerical data (values)to cartegorical data (values)

```python
def cp(row):
    if row == 0:
        return ' typical angina'
    elif row == 1:
        return 'atypical angina'
    elif row == 2:
        return 'non-anginal pain'
    elif row == 3:
        return 'asymptomatic'
```

```python
data['cp'] = data['cp'].apply(cp)
```

```python
data
```

```python
def slope(row):
    if row == 0:
        return 'upsloping'
    elif row == 1:
        return 'flat'
    elif row == 2:
        return 'downsloping'
```

```python
data['slope'] = data['slope'].apply(slope)
```

```python
data
```

```python
def thal(row):
    if row == 1:
        return 'normal'
    elif row == 2:
        return 'fixed defect'
    elif row == 3:
        return 'reversable defect'
```

```python
data['thal'] = data['thal'].apply(thal)
```
